```python
#インポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

#データフレーム作成
df = pd.read_csv('sample_pandas_6.csv')
cat_df = pd.read_csv('category.csv')
df = pd.merge(df, cat_df[["商品番号", "カテゴリー"]], how='inner', on="商品番号")

#カテゴリー列の要素(弁当、デザートなど)の出現頻度をカウントして、その結果を棒グラフで表示する
cat_bar = df["カテゴリー"].value_counts()
cat_bar.plot(kind='bar', xlabel="", title="カテゴリー別の発注頻度")
plt.show()

#商品番号列の商品番号ごとに、注文数の基本統計量を求める
shouhin = df.groupby("商品名")["注文数"].describe()
print(shouhin.round(2))
```
